```python
import pandas as pd 
import numpy as np
import scipy as scp
import matplotlib as mlp
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
from sklearn import preprocessing
import xgboost as xgb
from sklearn.cross_validation import train_test_split
```

```python
df=pd.read_csv('data_stocks.csv')
```

```python
df.head()
```

```python
DATE=df['DATE']
```

```python

date=DATE
a=[int(str(i)[0:4]) for i in date]
b=[int(str(i)[4:7]) for i in date]
c=[int(str(i)[7:]) for i in date]
df1=pd.DataFrame()
df1['a']=a
df1['b']=b
df1['c']=c
df1.dtypes
```

```python
for i in df.columns[1:]:
   Y=df[i]
   X_train, X_test, y_train, y_test = train_test_split(df1, Y, test_size=0.24, random_state=42)
               
               # use DMatrix for xgbosot
   import xgboost as xgb
           
           
   bst = xgb.XGBRegressor(max_depth=4, learning_rate=0.1, n_estimators=1000, silent=True, objective='reg:linear', )
   bst.fit(X_train,y_train)
   best_preds = bst.predict(X_test)
   print(i,r2_score(y_test,best_preds),mean_absolute_error(y_test,best_preds))
```
